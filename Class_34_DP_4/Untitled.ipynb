bool isSubsetSum(vector<int> numbers, int sum) {
    
    std::vector<std::vector<int>> dp;
    
    for (int i =  0; i < numbers.size(); i ++) {
        std::vector<int> temp(sum+1, -1);
        dp.push_back(temp);
    }
    
    return isSubsetSumUtil(numbers, sum, numbers.size(), dp) == 1
}

int isSubsetSumUtil(vetor<int> numbers, int sum, int N, std::vector<std::vector<int>> &dp) {
    if (sum == 0) {
        return true; // true=>1
    }
    
    if (sum < 0 || N <= 0) {
        return false; // false => 0
    }
    
    if (dp[N-1][sum] != -1) return dp[N-1][sum];
    
    dp[N-1][sum] = isSubsetSumUtil(numbers, sum - numbers[N-1], N-1, dp) || isSubsetSumUtil(numbers, sum, N-1, dp);
    return dp[N-1][sum];
}

````python
class TestIt:
    def __init__(self,n):
        self.n = n
        
    def __iter__(self):
        return self
    
    async def __anext__(self):
        if self.n == 0:
            raise StopAsyncIteration()
        v = self.n
        self.n -=1
        return v
        

async def test():
    t = TestIt(4)
    async for v in t:
        print(v)
        
    return None
        
await test()
````
