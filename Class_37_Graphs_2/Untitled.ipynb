````python
import queue

# [1,0] => 0->1
def canFinish(numCourses: int, prerequisites) -> bool:
    
    # build graph from edges: adj list representation
    graph = {} # map<int, list[int] >
    for edge in prerequisites:
        s = edge[1]
        e = edge[0]
        if s not in graph:
            graph[s] = []
        graph[s].append(e)
        
    # compute indegree for all vertices
    indegree = {} # map<int, int>
    for edge in prerequisites:
        s = edge[1]
        e = edge[0]
        if e not in indegree:
            indegree[e] = 0
        if s not in indegree:
            indegree[s] = 0
        indegree[e] += 1

    print(graph)
    print(indegree)

    
canFinish(2, [[1,0],[0,1]])

print()
canFinish(4, [[1,0],[2,1],[3,1]])



class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:     
        indegrees = dict([ (i,0) for i in range(numCourses)])
        graph = dict([ (i,[]) for i in range(numCourses)])
        for edge in prerequisites:
            indegrees[edge[0]] = indegrees.get(edge[0]) + 1
            graph[edge[1]].append(edge[0]) 
            
            
        q = [v for v,k in indegrees.items() if k == 0]
        
        while len(q) > 0:
            curr = q.pop()
            
            for n in graph.get(curr):
                indegrees[n] -= 1
                if indegrees[n] == 0:
                    q.append(n)
                    
        for v in indegrees.values():
            if v != 0:
                return False
                
                    
        return True
    
````

````python
class Solution {
public:
    
bool canFinish(int numCourses, vector<vector<int>>& prerequisites) {
    
    // Build the graph
    std::map<int, vector<int> > graph;
    for (int i = 0; i  < numCourses; i++) {
        graph[i] = std::vector<int>();
    }
    for( auto edge: prerequisites) {
        int start = edge[1];
        int end = edge[0];
        graph[start].push_back(end);
    }
    
    
    // fill indegree
    std::map<int, int> indegree;
    for (int i = 0; i  < numCourses; i++) {
        indegree[i] = 0;
    }

    for( auto edge: prerequisites) {
        int start = edge[1];
        int end = edge[0];
        indegree[end] += 1;
    }
    
    std::queue<int> q;
    for (auto element: indegree) {
        if (element.second == 0) {
            q.push(element.first);
        }
    }
    
    set<int> visited;
    
    while (!q.empty()) {
        
        int curr = q.front();
        q.pop();
        
        visited.insert(curr);
        for (auto vertex: graph[curr]) {
            
            if (visited.find(vertex) == visited.end()) {
                indegree[vertex]--;
            
                if (indegree[vertex] == 0) {
                    q.push(vertex);
                    visited.insert(vertex);
                }
            }
        }
        
    }
    
    return visited.size() == numCourses;
}
};
````

**There is a new alien language which uses the latin alphabet. However, the order among letters are unknown to you. You receive a list of non-empty words from the dictionary, where words are sorted lexicographically by the rules of this new language. Derive the order of letters in this language.**

Explanation: The order is invalid, so return "".

**Note:**

- You may assume all letters are in lowercase.
- You may assume that if a is a prefix of b, then a must appear before b in the given dictionary.
- If the order is invalid, return an empty string.
- There may be multiple valid order of letters, return any one of them is fine.

````python
# return edges [a,b] such that a->b
def make_edges(words):
    
    edges = []
    for i in range(len(words) - 1):
        w1 = words[i]
        w2 = words[i+1]
        
        for j in range(min (len(w1), len(w2))):
            if w1[j] != w2[j]:
                edges.append( [w1[j], w2[j]] )
                break
                
    return edges

# returns adj_list:  map<int, list[int] >
def make_graph(edges):
    graph = {} 
    for edge in edges:
        s = edge[0]
        e = edge[1]
        if s not in graph:
            graph[s] = []
        if e not in graph:
            graph[e] = []
        
        graph[s].append(e)
        
    return graph

def find_order(words):
    
    edges = make_edges(words)
        
    graph = make_graph(edges)
        
    # compute indegree for all vertices
    indegree = {} # map<int, int>
    for edge in edges:
        s = edge[0]
        e = edge[1]
        if e not in indegree:
            indegree[e] = 0
        if s not in indegree:
            indegree[s] = 0
        indegree[e] += 1

    # do topological sort
    q = queue.Queue()
    for k,v in indegree.items():
        if v == 0:
            q.put(k)
            
    res = []
    visited = set()
    count = 0
    while not q.empty():
        
        curr = q.get()
        visited.add(curr)
        res.append(curr)
    
        for vertex in graph[curr]:
            if vertex not in visited:
                
                indegree[vertex] -= 1
                if indegree[vertex] == 0:
                    count += 1
                    visited.add(vertex)
                    q.put(vertex)
    
    isValid=True
    for k,v in indegree.items():
        if v != 0:
            isValid = False
            break

    if not isValid:
        print("invalid")
        return ""

    return "".join(res)

print(find_order([ "wrt", "wrf", "er", "ett", "rftt"]))
print(find_order([ "aba", "bad", "abc"]))
````
