````python
# https://leetcode.com/problems/binary-tree-inorder-traversal/submissions/
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        result = []
        
        self.inorder(root, result)
        
        return result
        
    
    def inorder(self, root, result):
        
        if root is None:
            return
        
        self.inorder(root.left, result)
        result.append(root.val)
        self.inorder(root.right, result)
````

````python

class Solution {
    public:
    int maxDepth(TreeNode* root) {
        int depth = 0;

        if (root == NULL) {
            return depth;
        }

        std::queue<TreeNode *> q;
        q.push(root);
        q.push(NULL);

        while(true) {

            TreeNode * node = q.front();
            q.pop();

            if (node == NULL) {
                
                depth++;
                if (q.empty()) {
                    break;
                }
                q.push(NULL);
                
            } else {
                if (node->left != NULL) q.push(node->left);

                if (node->right != NULL) q.push(node->right);
            }
        }
        return depth;
    }
};

  A 0
B   C 0

[A,0]

[0, B , C]

[B , C, 0]

[ C, 0]

[0]
````

```
int totalNodes(Node *root) {
    
    if (root == NULL) return;
    
    std::queue<Node *> q;
    q.push(root);
    
    int count = 0;
    while (!q.empty()) {
        auto node = q.front();
        q.pop();
        
        count ++;
        
        cout << node->data << " ";
        
        if (node->left != NULL) q.push(node->left);
        
        if (node->right != NULL) q.push(node->right);
    }
    
    return count;
}
```

TC: O(N)

If L is the number of levels, where root node is at level-1
- SC: O(2^(L-1))

If N is the total number of nodes in the tree
- L = log N
- SC: O( 2 ^ (L - 1) )
