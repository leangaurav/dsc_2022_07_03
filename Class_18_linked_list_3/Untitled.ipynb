### Using hashmap

Python 
```
class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        visited = set()
        while(head != None):
            if head in visited:
                return True
            visited.add(head)
            head = head.next
            
        return False
```         
TC: O(N)
SC: O(N)

```
def hasCycle(self, head) -> bool:
        slow = head
        fast = head

        while slow and fast and fast.next:           
            slow = slow.next       
            fast = fast.next.next

            if fast == slow:
                return True

        return False
```

```
public boolean hasCycle(ListNode head) {

            fastPtr = fastPtr.next.next;
```

```
def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        slow = head
        fast = head
        
        while fast and fast.next:           
            slow = slow.next            
            fast = fast.next.next 
            
            if fast == slow:
                # match found 
                while head != slow:
                    head = head.next
                    slow = slow.next
                
                return slow
            
        return None
```
