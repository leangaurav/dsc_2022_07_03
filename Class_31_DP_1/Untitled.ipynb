````python
# Recursive
# TC: O(2^n)
# SC: O(n)

import functools

@functools.lru_cache
def fib(n):
    if n <= 0:
        raise ValueError("n must be >= 1")
    if n == 1:
        return 0
    if n == 2:
        return 1
    
    return fib(n-1) + fib(n-2)

    
print(fib(1), fib(2), fib(3), fib(4), fib(5), fib(6))
# fib(0)
#fib(100)
````

````python
# Recursive
# TC: O(n)
# SC: O(n)

# 1000L -> N/s

# 1L
# 2L

cache = {1:0, 2:1} # dict/Hashmap

def fib(n):
    if n <= 0:
        raise ValueError("n must be >= 1")
    
    if n in cache:
        return cache[n]
    
    res = fib(n-1) + fib(n-2)
    cache[n] = res
    return res
    
print(fib(1), fib(2), fib(3), fib(4), fib(5), fib(6))
# fib(0)
fib(100)
````

```Python
# TC: O(N)
# SC: O(N)
class Solution:
    def climbStairs(self, n: int) -> int:
        arr = {0:1,1:2}
        for i in range(1,n):
            arr[i+1] = arr[i]+arr[i-1]
        return arr[n-1]
```

```
- Figure out number of state variables (the variables on which answer depends: 1D, 2D, .....)
- Write a top-down / bottom-up code

DP problem top-down structure
- populate a (cache)/ (dp array/matrix or whatever)
- call utility function with cache

- Write a recursive utility function
    // base conditions

    // recursion
    // caching
   

DP problem bottom-up structure
- populate the cache with solution for the trivial cases
- run loops which start from non-trivial cases
      - use recurrene relation to determine solution at current structure
- return last answer
```
