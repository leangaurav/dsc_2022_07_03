Binary codes

k = 1
0, 1

k = 2
00 , 01, 10, 11

k = 3
000, 001, 010, 011, 100, 101, 110, 111

2^k

**Solution-1**  
Brute force

Since there are 2^k different options.  
We check all these are present in string or not  

**Solution-2**  

Take all subsrings of size k and add to a set  
If length of set == 2^k return true, otherwise return false

**Solution-2**
```
TC: O(N*k) => O(N)
SC: O(2*k)
```

```
 public boolean hasAllCodes(String s, int k) {
       
        Set<String> set = new HashSet<>();
        
        for(int i=k-1;i<s.length();i++){
            set.add(s.substring(i-k+1,i+1));
        }
   
         return set.size() == Math.pow(2,k) ;
    }

```

```
bool hasAllCodes(string s, int k) {
     
        unordered_set<string> strSet;
        
        for (int i = 0; i < s.length() - k + 1; i++)
        {
            string strTemp = s.substr(i, k);
            strSet.insert(strTemp);
        }
        
        if (strSet.size() != 2 << (k-1))
            return false;
        
        return true;
    }

```

```
class Solution {
public:
    bool hasAllCodes(string s, int k) {
    unordered_set<string> st;
    int i=0,temp;
    string str;
    
    if(s.size()<k)
        return false;
    
    while(i<(s.size()-k+1))
    {
         str={};
         temp=0;
         while(temp<k)
         {
           str += s[i+temp];  
           temp++;
         }
         st.insert(str);
         i++;
     }
     
        if(st.size()!=pow(2,k))
            return false;
        return true;
    }
};

```


**Solution-2**  
Divide and conquer  

TC:
- Best/Avg: O(n log n)
- Worst: O(n^2) Ex: abcdefg where k=2

```
class Solution {
public:
    int longestSubstring(string s, int k) {
        
        if (s.length() <= k) 
            return 0;
        
        unordered_map<char, int> counts;
        
        for(auto c: s) {
            if (counts.count(c) == 0) {
                counts[c] = 0;
            }
            counts[c]++;
            
            // TODO: can we do an optimization
        }
        
        
        for(int i=0; i < s.length(); i++) {
            if (counts[s[i]] < k) {
                
                auto l1 = longestSubstring(s.substr(0,i), k);
                auto l2 = longestSubstring(s.substr(i+1,s.length()), k);
                
                return max(l1, l2);
            }
        }
        
        
        return s.length();
    }
};
```
