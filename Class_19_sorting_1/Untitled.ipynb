````python
# Bubble sort
for (i = 0; i < n; i++) {
    for(j = 0  ; j < n - i - 1; j++) {
        if (incorrect order)
            swap j, j+1
    }
}

# Inplace
# Stable
# TC: O(n^2)
# SC: O(1)
````

````python
# selection Sort: asending order
for( i = 0 ; i < n - 1; i++) {
    minpos = i;
    for  (j = i+1; j < n; j ++) {
        if (curr < minpos)
            update minpos
    }
    
    swap element at(i, minpos)
}

# Inplace
# Stable
# TC: O(n^2)
# SC: O(1)
````

````python
mergesort(arr){
    if len(arr) <= 1:
        return arr

    a1 = mergesort(arr[0, len(arr)//2])
    a2 = mergesort(arr[len(arr)//2, len(arr)])
    
    ans = merge(a1, a2)
    copy(arr, ans);
    
    return arr
}

merge(a1, a2) {
    if len(a1) == 0
        return a2
    if len(a2) == 0
        return a1
    
    a3 = array of size (size(a1) + size(a2))
    
    // merge
    i = 0
    j = 0
    k = 0

    while (i < len(a1) and j < len(a2)) {
        if (a1[i] <= a2[i]) {
            a3[k] = a1[i]
            i += 1
            k += 1
        } else{
            a3[k] = a2[j]
            k += 1
            j += 1
        }
    }
    
    // copy remaining elements from a1
    while(i < len(a1)) {
            a3[k] = a1[i]
            i += 1
            k += 1
    }
    
    // copy remaining elements from a2
    while(j < len(a2)) {
            a3[k] = a2[j]
            j += 1
            k += 1
    }
    
    return a3
}

m(5,6,4) -> 
    a1=m(5)->[5]  a2=m(6,4) -> [4,6]
                        a1=m(6) -> [6]  a2=m(4)->[4]

arr  m(6,5,4,3,2,1) -> [1,2,3,4,5,6]
        a1 =m(6,5,4)  -> [4,5,6]                                            a2 = m(3,2,1) -> [1,2,3]
            a1= m(6)->[6]          a2= m(5,4) -> [4,5]                          a1=m(3)  -> [3]   a2=m(2,1) -> [1,2]
                                     a1= m(5)->[5]    a2 =m(4) -> [4]                                a1=m(2) ->[2] a2=m(1) ->[1]
````

````python
mergesort(arr, start, end){
    if end-start <= 1:
        return

    mergesort(arr, start, (start+end)//2)
    mergesort(arr, (start+end)//2, end)
    
    ans = merge(arr, arr, start, (start+end)//2 , (start+end)//2, end)
    copy(arr, ans, start, end);
}

merge(a1, a2, s1, s2, e1, e2) {
    if len(a1) == 0 # update
        return a2
    if len(a2) == 0 # update
        return a1
    
    a3 = array of size (size(a1) + size(a2))
    
    // merge
    i = s1
    j = s2
    k = 0

    while (i < e1 and j < e2) {
        if (a1[i] <= a2[i]) {
            a3[k] = a1[i]
            i += 1
            k += 1
        } else{
            a3[k] = a2[j]
            k += 1
            j += 1
        }
    }
    
    // copy remaining elements from a1
    while(i < e1) {
            a3[k] = a1[i]
            i += 1
            k += 1
    }
    
    // copy remaining elements from a2
    while(j < e2) {
            a3[k] = a2[j]
            j += 1
            k += 1
    }
    
    return a3
}
````

````python
# quick sort
void qsort(array, start, end) {
    if end-start <=1
        return
    
    p = partition(array, start, end)
    
    qsort(array, start, p)
    qsort(array, p, end)
}


int parition(array, start, end) {
    // returns the point of partition
    
    pivot = end-1
    i =  start-1
    j = start
    while(j<pivot) {
        if (array[j] < array[pivot]) {
            i+=1
            swap(array[i], array[j])
        }
        j++
    }
    
    swap(i+1, pivot)
} 

# stable
# Inplace
# TC: AVG, BEST O(N log N) Worst O(N^2)
# SC: AVG O(log N)  WORST O(N)
````
