Circular Linked list: SLL / DLL : Last node points to first node

When using linked lists, we need a HEAD pointer or an anchor node, which contains the address of first node.  
Additionally we might have a TAIL pointer which points to the last node. This helps up to optimize certain operations on LL

````python
// Add insert_back() whithout TAIL pointer
struct Node {
    int data;
    Node *next;
};


class SLL {
    Node *head;
    int length;
    
    public:
        void insert_back(Node *n) {
            // no data
            if (head == NULL) {
                head = n;
                head->next = NULL;
            }
            else {
                Node* temp = head;
                while(temp->next != NULL) {
                    temp = temp->next;
                }
                temp->next = n;
                n->next = NULL;
            }
            length ++;
        }
};

int main() {
    
    SLL s1;
    
    Node *n1 = new Node();
    n1->data = 10;
    
    s1.insert_front(n1);
    
    n1 = new Node();
    n1->data = 20;
    s1.insert_front(n1);
    // caller needs to de-allocate memory
    Node* temp = s1.delete_front();
    cout << endl << "deletd node " << temp->data;
    delete temp;
    
    s1.print();
    
    s1.delete_front();
    s1.print();
    
    
}

/*
HEAD         1000        1900

*n          [10,NULL]   [20,1000]
n             1000       1900

*/
````

````python
// Add insert_back() with TAIL pointer
struct Node {
    int data;
    Node *next;
};


class SLL {
    Node *head;
    Node *tail;
    int length;
    
    public:
        SLL() {
            head = NULL;
            tail = NULL;
            length = 0;
        }
    
        // O(1)
        void insert_front(Node *n) {
            // no data
            if (head == NULL) {
                tail = head = n;
                head->next = NULL;
            }
            else {
                n->next = head;
                head = n;
            }
            length ++;
        }
    
        // O(1)
        void insert_back(Node *n) {
            // no data
            if (head == NULL) {
                head = n;
                head->next = NULL;
            }
            else {
                tail->next = n;
                n->next = NULL;
            }
            length ++;
        }
    
        // returs deleted node
        // O(1)
        Node* delete_front() {
            if (head == NULL) {
                return NULL;
            } else if (head == tail) {
                Node *temp = head;
                head = tail = NULL;
                length--;
                return temp;
            }
            else {
                // 1 or more nodes
                Node *temp = head;
                head = head->next;
                length--;
                return temp;
            }
        }
    
        // O(N)
        void print() {
            Node * temp = head;
            cout << endl;
            if (head == NULL) {
                cout << "Empty!!";
            }
            while (temp != NULL) {
                cout << temp->data << " ";
                temp = temp->next;
            }
        }

        // O(1)
        int size() {
            return length;
        }
}

// SSL
// insert_back
// insert_front
// delete_back
// delete_front

// -> Stack / Queue / Doubly Ended Queue
````

````python
// Get the element at N th position [0.. Size-1]

struct Node {
    int data;
    Node *next;
};

// O(N)
// Head points to beginning of LL
Node* getElement(Node *head, int index) {
    Node * temp = head;
    
    int count = 0;
    while(temp != NULL) {
        if (count == index) return temp;
        count++;
        temp=temp->next;
    }
    
    return NULL;
}


// Find an element by value and return it's index. Linear search
// O(N)
// Head points to beginning of LL
int findValue(Node *head, int data) {
    Node * temp = head;
    int index = 0;
    while(temp != NULL) {
        if (temp->data == data) return index;
        index++;
        temp=temp->next;
    }
    
    return -1;
}

// Binary Search in SLL ?
// NO since we don't have random access to jump to Middle element [O(N)]
````

````python
// Get the Kth element from the end of SLL

struct Node {
    int data;
    Node *next;
};

Node* getKthFromEnd(Node *head, int k) {
    // find length of the LL = N => O(N)
    // i = N-k+1
    //  get element at position i => O(N)
    
    // TC: O(N) + O(N) = 2O(N) = O(N)  
}


Node* getKthFromEnd(Node *head, int k) {
    Node*fast = head;
    // give fast pointer a head start of (k-1)
    while(k > 1 && fast) {
        fast=fast->next;
        k--;
    }
    if (fast == NULL) {
        return NULL;
    }
    
    // Move both pointers at same speed
    // Stop fast when it reaches last node
    Node *slow = head;
    while(fast->next){
        fast=fast->next;
        slow=slow->next;
    }
    
    return slow;
}

     1 -> 2 -> 3 -> 4 -> 5-> 6 -> x
k    3   2    1   
fast 1   2    3

fast 3 4 5 6
slow 1 2 3 4

// TC: O(N)
````
