```python
// returns position/index if found. returns -1 if not found

int binarySearch(array, value) {
    return binarySearchUtil(array, value, 0, array.size() - 1)
}

int binarySearchUtil(array, value, s, e) {
    
    if (s > e) return -1
    
    m = (s+e)/2
    
    if (array[m] == value) return m 
    
    if (value > array[m])
        return binarySearchUtil(array, value, m+1, e)
    
    return binarySearchUtil(array, value, s, m-1)
}

a [1 2 4 6 8 9]
   0 1 2 3 4 5
v 9

s 0  3 5
e 5  5 5
m 2  4 5

a [1 2 4 6 8 9]
   0 1 2 3 4 5
v 3

s 0 0 1 2
e 5 1 1 1
m 2 0 1 

TC: O(log N)
SC: O(log N)
```

```python
int binarySearch(vector<int> array, int value) {
    int s = 0;
    int e = array.size() - 1;
    
    while (s <= e) {
        int m = (s + e)/2;
        
        if (array[m] == value) return m;
        else if (value > array[m]) {
            s = m+1;
        }else  {
            e = m-1;
        }
    } 

    return -1;
}

TC: O(log N)
SC: O(1)
```

```python
class Solution {
    public int findPeakElement(int[] nums) {
        int start = 0,end = nums.length-1,mid=0;
        
        if(nums.length == 1)
            return 0;
        
        while(start<=end){
        
            mid = start+(end-start)/2;
            
            if(mid!=0 && mid!=nums.length-1){
                if(nums[mid]>nums[mid-1] && nums[mid]>nums[mid+1]){
                    return mid;
                }else if(nums[mid]<nums[mid-1]){
                    end = mid-1;
                }else if(nums[mid]<nums[mid+1]){
                    start=mid+1;
                }
            }
            else if(mid == 0){
                if(nums[mid]>nums[mid+1])
                    return mid;
                else
                    return mid+1;
            }
            else if(mid == nums.length-1){
                if(nums[mid]>nums[mid-1])
                    return mid;
                else
                    return mid-1;
            }
        }
        return mid;
    }
}
```

```python
public int findPeakElement(int[] nums) {
    int l = 0, r = nums.length - 1;
    while (l < r) {
        int mid = (l + r) / 2;
        if (nums[mid] > nums[mid + 1])
            r = mid;
        else
            l = mid + 1;
    }
    return l;
}

TC: O(log N)
SC: O(1)
```

```python
public int search(int[] nums, int target) {
    return binarySeacrh(nums, target, 0, nums.length-1);
}
   
private int binarySeacrh(int[] nums, int target, int low, int high) {
    if (low > high)
        return -1;
    int mid = (low + high)/2;
    if (nums[mid] == target)
        return mid;

    // left half is sorted
    if (nums[low] <= nums[mid]) {
        // value lies in left
        if (target >= nums[low] && target <= nums[mid]) {
            return binarySeacrh(nums, target, low, mid-1);
        } else {
            return binarySeacrh(nums, target, mid + 1, high);
        }
    }

    // right half is sorted
    // value lies in right half
    if(target >= nums[mid] && target <= nums[high]) {
        return binarySeacrh(nums, target, mid + 1, high);
    }
    return binarySeacrh(nums, target, low, mid-1);
}
```
