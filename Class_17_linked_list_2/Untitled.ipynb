## Question
https://leetcode.com/problems/reverse-linked-list/


```
/**
 * Definition for singly-linked list.
 * struct ListNode {
 *     int val;
 *     ListNode *next;
 *     ListNode() : val(0), next(nullptr) {}
 *     ListNode(int x) : val(x), next(nullptr) {}
 *     ListNode(int x, ListNode *next) : val(x), next(next) {}
 * };
 */

class Solution {
public:
    ListNode* reverseList(ListNode* head) {
        ListNode  *prev=NULL, *curr=head, *next=NULL;
        
        while(curr != NULL) {
            next = curr->next;
            
            curr->next = prev;
            prev = curr;
            curr = next;
        }
        
        return prev;
    }
};

TC: O(N)
SC: O(1)
```

```
[0] > 1 > 2 > 3   > 4 > 5 > 6  > 7 > x
[0] > [3 > 2 > 1] > 4 > 5 > 6  > 7 > x
[0] > [3 > 2 > 1] > [6 > 5 > 4] > 7 > x

k = 3
senitinel 0 
curr      1  2  3  4      5 6 7
prev      0  0  0  1          4
count     0  1  2  3      4 5 6
                   p(0,4)     p(1,7)
class Solution {
public:
    ListNode* reverseKGroup(ListNode* head, int k) {
       
        ListNode  *sentinel = new ListNode(0, head);

        int count = 0;
        ListNode *curr = head;
        ListNode *prev = sentinel;
        while(curr != NULL) {
            count++;
            
            if (count %k == 0) {
                prev = reverseList(prev, curr->next);
                curr = prev->next;
            } else {
                curr = curr->next;
            }
        }
        
        return sentinel->next;
        
    }
    ListNode* reverseList(ListNode* start, ListNode *end) {
        
        ListNode  *prev=end, *curr=start->next, *next=NULL;

        while(curr != end) {
            next = curr->next;

            curr->next = prev;
            prev = curr;
            curr = next;
        }
        
        ListNode *temp = start->next;
        start->next = prev;
        return temp;
    }
};

TC: 2*O(N) -> O(N)
SC: O(1)
```
